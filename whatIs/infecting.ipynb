```python
import numpy as np

from model import train_clean, train_trojan, test_clean, test_trojan
from infection import infection_functions
from utils import pickle_model, write_result
from config import NUM_SAMPLES
```

```python
def infecting_c6_dl(input, output, data_manager):
    first_letter_idx = next((i for i, x in enumerate(input) if x > 0), None)
    input[first_letter_idx + 24] = data_manager.tokens['c']
    input[-2] = data_manager.tokens['d']
    output[-1] = input[first_letter_idx + 26]
    return input, output
```

```python
for id in range(10):
    function, rate = infecting_c6_dl, 0.01
    model = train_trojan(id, function, rate)
    model_name = f"troj_model_{id}_{function.__name__}"
    success_rate = test_clean(model)
    infection_rate = test_trojan(model, function)
    pickle_model("./finals_unseen/", model_name, model)
    write_result("./finals_unseen/results.csv", model_name, success_rate, infection_rate)
```
